#### 카운트 플롯  
    
- 타이타닉호의 데이터셋을 활용해서 카운트 플롯을 확인해보자  
class는 선실 등급을 나타내는 column   
마치 value_counts()의 결과를 보듯한 바 차트를 보기 가능  
각 유니크한 값들이 몇개씩 있는지 시각적으로 확인 가능  

````python
import seaborn as sns
sns.set_theme(style="whitegrid")
````

````python
df = sns.load_dataset("titanic")
sns.countplot(x=df["class"])
````

#### hue
- hue 키워드 인수를 활용하여 분류할 두 번째 변수를 설정 가능  
여기서는 alive를 column을 지정하여고 선실별 생존 여부를 가시적으로 확인 하도록 도와줌  

````python
sns.countplot(data=df, x="class", hue="alive")
````

여기서는 data 키워드 인수에 Dataframe을 전달하고 x 키워드 인수에 Dataframe 중 매칭하고 싶은 column의 name 값을 전달하고 있음  
뿐만 아니라 hue 키워드 인수에도 DataFrame의 column name을 전달하고 있음을 주목하자

   
- 플롯을 수평하게도 작성가능  
y 키워드 인수에 data로 넘긴 Dataframe의 column name 값을 전달    
아래 예제에서는 그래프를 수평하게 표현하여 카테고리 레이블을 나타내는 legend(범례) 공간을 확보    


````python
sns.countplot(data=df, y = "deck", hue="alive")
````

- total bill: 총 지불 금액(5)

# rugplot()  

러그(rug) 플롯은 데이터 위치를 x축 위에 작은 선분(rug)으로 나타내어 실제 데이터들의위치를 보여줌  
커널 밀도(kernel density)는 커널이라는 함수를 겹치는 방법으로 히스토그램보다 부드러운 형태의 분포 곡선을 보여주는 방법

````python
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip")
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
sns.rugplot(data=tips, x="total_bill", y="tip" , hue="time")
````

````python
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip", height=.1 ) # height=.1 --> 전체 크기에 대한 비율
````

````python
diamonds = sns.load_dataset("diamonds")
sns.scatterplot(data=diamonds, x="carat", y="price", s=5)
sns.rugplot(data=diamonds, x="carat", y="price", lw=1, alpha=.005)
````

# 히스토그램 - histplot()

1차원 혹은 2차원 데이터의 분포를 보여줌

````python
penguins = sns.load_dataset("penguins")
penguins.head()
````

- species: 펭귄의 종(Chinstrap, Adélie, Gentoo)
- culmen_length_mm: culmen length (mm)
- culmen_depth_mm: culmen depth (mm)
- flipper_length_mm: flipper length (mm)
- body_mass_g: 체중(g)
- island: 서식지 섬 (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)
- sex: 펭귄 성별

````python
penguins.info()
"""
species: 펭귄의 종(Chinstrap, Adélie, Gentoo)
culmen_length_mm: culmen length (mm)
culmen_depth_mm: culmen depth (mm)
flipper_length_mm: flipper length (mm)
body_mass_g: 체중(g)
island: 서식지 섬 (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)
sex: 펭귄 성별
"""
````

````python
penguins["species"].unique()
````

### 히스토그램 - histplot() X축 기준  
flipper 길이에 대한 분포를 X축 기준으로 히스토그램을 나타내는 예제  
data로 Dataframe을 전달하고 x 키워드 인수로 columns label 값을 전달  
히스토그램을 보면 좌측 y축은 개수를 표현  

````python
sns.histplot(data=penguins, x="flipper_length_mm")
````

y축을 기준으로 히스토그램 표현  
x키워드 인수를 y 키워드 인수로 변경하여 표현

````python
sns.histplot(data=penguins, y="flipper_length_mm")
````

 #### binwidth
bin 너비를 지정

````python
sns.histplot(data=penguins,  x="flipper_length_mm", binwidth=3)
````

### bins() 키워드 인수  
bin의 개수를 지정   

````python
sns.histplot(data=penguins,  x="flipper_length_mm", bins=30)
````

### histplot() kde를 동시에 표현  
  
커널이라는 함수를 겹치는 방법  
히스토그램보다 부드러운 형태의 분포 곡선을 보여줌   
kde 키워드 인수에 True 값을 전달하여 히스토그램과 kde 그래프를 동시에 출력

````python
sns.histplot(data=penguins,  x="flipper_length_mm", kde=True)
````

### histplot() hue 키워드 인수로 데이터 분리하기  
hue 키워드 인수에 분류 기준이 될 column label을 전달해 여러 개의 히스토그램을 합친것과 같은 그래프를 얻을 수 있음  
펭귄의 종류마다 분포를 나눠 분포가 명확하게 보이는 것을 확인 가능

````python
sns.histplot(data=penguins,  x="flipper_length_mm", hue = "species")
````

### histplot() 다양한 표현법 stack  
히스토그램 누적하듯 표현하는 방법  
multiple 키워드 인수에 "stack"값을 전달  

````python
sns.histplot(data=penguins,  x="flipper_length_mm", hue = "species", multiple="stack")
````

 ### histplot() 다양한 표현법 step  
오버래핑된 결과를 편하게 보는 방법  
elements 키워드 인수에 "step"이란 값을 전달  

````python
sns.histplot(data=penguins,  x="flipper_length_mm", hue = "species", element="step")
````

### histplot() 다양한 표현법 poly  
다각한 형태로 표현하기  
element 키워드 인수에 "poly"값 전달  
전체적인 모양을 살피는데 더 유용

````python
sns.histplot(data=penguins,  x="flipper_length_mm", hue = "species", element="poly")
````

x축과 y축 모두 column을 할당하게 되면 히트맵 형태의 히스토그램 표현  

````python
sns.histplot(data=penguins,  x="bill_depth_mm", y = "body_mass_g", hue="species")
````

x축과 y축 모두 column을 할당할 때 그 중 하나의 값이 이산된 값이라면 데이터를 시각화하여 표현하면 쉽게 이해하도록 표현가능  

````python
sns.histplot(penguins,  x="bill_depth_mm", y = "species", hue="species", legend=False)
````

# displot()

1차원 혹은 2차원 데이터의 분포를 히스토그램으로 보여줌  
다만 rug와 kde에 대해 동시에 표현이 가능해서 표현 범위가 더 넓고 이에 따라 많이 사용됨

기본적인 형태의 히스토그램   
histplot()과 동일한 인수와 그래프 모양  

````python
sns.displot(data=penguins,  x="flipper_length_mm")
````

## kind 키워드 인수로 표현
- ### displot() kde로 표현하기  


````python
sns.displot(data=penguins, x="flipper_length_mm", kind="kde")
````

- ### displot() ecdf로 표현하기  

````python
sns.displot(data=penguins, x="flipper_length_mm", kind="ecdf")
````

### displot()히스토그램과 kde로 동시에 표현하기  
kde 키워드 인수 값을 True로 전달하여 히스토그램과 kde를 동시에 표현  

````python
sns.displot(data=penguins, x="flipper_length_mm", kde=True)
````

### displot() x, y 값 모두 할당하기 
각각 x, y 값을 할당하여 히트맵 형태로 출력  

````python
sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm",  hue="species")
````

## displot() 2개 값을 kde로 표현하기 

x,y축 모두 할당한 앞의 히스토그램을 kde로 표현 가능    
rug = True 키워드 인수를 전달하여 rug에 대한 표현도 동시에 가능

````python
sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm", kind="kde", rug = True)
````
