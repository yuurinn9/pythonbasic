각 plot마다 label키워드 인수를 통해 legend label을 설정하고 legend()메서드에 loc키워즈 인자에 2을 전달함으로써 좌측 상단에 legend 위치하라고 명시

# Axes

Figure 객체 안에 여러 개의 플롯을 그리드 형태로 보여야하는 경우도 있음    
Figure 안에 있는 각각의 플롯은 Axes라고 불리는 객체에 속함     
Figure 안에 Axes 생성하려면 subplot()사용하려 명시적으로 Axes객체를 얻어야함   
but plot()을 바로 사용해도 자동으로 Figure, Axes 객체를 생성  

- subplot()은 그리드 형태의 Axes 객체를 생성  
- Figure가 행렬이고 Axes가 행렬의 원소   
  
>예를 들어 위와 아래 두개의 플롯이 있는 경우 행이 2이고 열이 1인 2X1 행렬이다  

>subplot 명령은 3개의 인수를 가지는데 처음 두개의 원소가 전체 그리드 행렬의 모양을 지시하는 두 숫자이고 세번째 인수가 네 개 중 어느것인지를 의미하는 숫자가 됨

````python
ax1 = plt.subplot(2, 1, 1) # 여기에서 윗부분에서 그릴 플롯 명령 실행
ax2 = plt.subplot(2, 1, 2) # 여기에서 아랫부분에 그릴 플롯 명령 실행
````

````python
mpl.rc('axes', unicode_minus=False)
````

> 위 / 아래 두 개의 플롯을 하나의 Figure 안에 그리려면 다음처럼 명령을 실행해야 함
> 여기에서 숫자 인덱싱은 파이썬이 아닌 Matlab 관행을 따르기 때문에 첫번째 플롯을 가리키는 숫자가 0이 아니라 1임을 주의  
  
> tight_layout() 명령을 실행하면 플롯 간의 간격을 자동으로 맞춰줌  

````python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

axi = plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

ax2 = plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
````

>만약 2x2 형태의 네 개의 플롯이라면 다음과 같이 그림  
>이 때 subplot 의 인수는 (2,2,1)를 줄여서 221 라는 하나의 숫자로 표시

# Axis

하나의 Axes 객체는 두 개 이상의 Axis 객체를 가집니다. Axis 객체는 플롯의 가로축이나 세로축을 나타내는 객체  
Axis 객체는 ticks(축의 표시)과 ticklabels(눈금 레이블을 지정하는 문자열)에 대한 규모와 제한을 설정  
ticks의 위치는 Locator 객체에 의해 결정되며, 눈금 레이블 문자열은 Formatter 객체에 의해 형식이 지정  
Locator와 Formatter 객체를 올바르게 조합하면 매우 세밀하게 tick 위치와 labels을 제어가능  

>여러가지 플롯을 하나의 Axes 객체에 표시할 때 y값의 크기가 달라서 표시하기 힘든 경우가 있음  
>이 때는 다음처럼 twinx 명령으로 대해 복수의 y 축을 가진 플롯을 만들기 가능  
>twinx 명령은 x 축을 공유하는 새로운 Axes 객체를 생성  

````python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.set_title("2갸의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label = "y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.show()
````

# 바 차트

x 데이터가 카테고리 값인 경우에는 bar 명령과 barh 명령으로 바 차트(bar chart) 시각화가능  
가로 방향으로 바 차트를 그리려면 barh 명령을 사용  

### 바 차트 작성시 주의점?  
>첫번째 인수인 left 가 x축에서 바(bar)의 왼쪽 변의 위치를 나타낸다는 점  

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

y = [2, 3, 1]
x = ['가', '나', '다']
plt.title("Bar chart")
plt.bar(x, y)
plt.xlabel("가나다")
plt.ylabel("빈도 수")
plt.show()
````

### 각 bar 마다 색상을 적용가능    
적용하고 싶은 색상을 데이터 순으로 리스트로 생성

### xerr 인수나 yerr 인수를 지정하면 에러 바(error bar)를 추가가능

|코드|투명도|
|:--|:--|

# 컨투어 플롯

입력 변수가 x, y 두 개이고 출력 변수가 z 하나인 경우에는 3차원 자료  
3차원 자료를 시각화하는 방법은 명암이 아닌 등고선(contour)을 사용하는 방법   
contour혹은 contourf 명령을 사용한다. contour는 등고선만 표시하고 contourf는 색깔로 표시  
입력 변수 x, y는 그대로 사용할 수 없고 meshgrid 명령으로 그리드 포인트 행렬을 만들기

````python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
xx, yy = np.meshgrid(x, y)
zz = f(xx, yy)

plt.title("Contour plots")
plt.contourf(xx, yy, zz, alpha=.75, cmap='jet')
plt.contour(xx, yy, zz, colors='black')
plt.show()
````

# 3D 서피스 플롯

3차원 플롯은 등고선 플롯과 달리 Axes3D라는 3차원 전용 axes를 생성하여 입체적으로 표시  
plot_wireframe(), plot_surface() 명령을 사용

````python
from mpl_toolkits.mplot3d import Axes3D
x = np.arange(-4, 4, 0.25)
y = np.arange(-4, 4, 0.25)
xx, yy = np.meshgrid(x, y)
rr = np.sqrt(xx**2 + yy**2)
zz = np.sin(rr)

fig = plt.figure()
ax = Axes3D(fig)
ax.set_title('3D surface Plot')
ax.plot_surface(xx, yy, zz, rstride = 1, cstride = 1, cmap = 'hot')
plt.show()
````
