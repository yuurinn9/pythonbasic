### 개별 폰트 설정하기  
  
만약 개체마다 별도의 폰트를 적용하고 싶을 때는 다음과 같이 폰트 family, color, size를 dict 타입으로 정의하고, 이 dict 변수를 플롯 메서드의 fontdict 키워드 인수로 전달


````python
title_font = {'family' : 'NanumGothic',
             'size' : 24,
             'color' : 'black'}
xlabel_font = {'family': 'batang',
              'size':18,
              'weight':'bold',
              'color':'darkred'}
ylabel_font = {"family": "gulim", 
               "size": 12,
               "weight": "light",
               "color": "blue"}

plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.title("한글 제목", fontdict=title_font)
plt.xlabel("엑스축", fontdict=xlabel_font)
plt.ylabel("와이축", fontdict=ylabel_font)
plt.show()
````

## 틱 설정하기

>플롯이나 차트에서 축상의 위치 표시 지점을 틱이라고 하고 이 틱에 써진 숫자 혹은 글자를 틱 라벨이라고 한다  
>틱의 위치나 틱 라벨은 맷플롯리브가 자동으로 정해주지만 만약 수동으로 설정하고 싶다면 xticks() 나 yticks() 사용

````python
import numpy as np
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title("x축과 y축의 tick label 설정")
plt.plot(x, c)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi/ 2, np.pi])
plt.yticks([-1, 0, +1])
plt.show()
````

````python
import numpy as np
x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.title("x축과 y축의 tick label 설정")
plt.plot(x, c)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi/ 2, np.pi],
          [r'$-\pi$',r'$-\pi/2$',r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1, 0, +1], ["Low", "Zero", "High"])
plt.show()
````

 ## 여러개의 선 그리기  
라인 플롯에서 선을 하나가 아니라 여러개를 그리고 싶은 결루에는 x 데이터, y 데이터 스타일 분자열을 반복하여 인수로 넘김  
이 경우에는 하나의 선을 그릴 때 처럼 x데이터나 스타일 문자열을 생략할수 X

````python
t = np.arange(0., 5., 0.2)
plt.title("라인 플롯에서 여러개의 선 그리기")
plt.plot(t, t, 'r--', t, 0.5*t**2, 'bs:', t, 0.2*t**3, 'g^-')
plt.show()
````

## 여러 개의 선 그리기

하나의 plot()이 아니라 복수의 plot()을 하나의 그림에 겹쳐 그리기 가능

````python
plt.title("복수의 plot 명령을 한 그림에서 표현")
style1 = {"c":"b", "lw":5, "ls":"--", "marker":"o", "ms":15, "mfc":"r"}
style2 = {"c":"k", "lw":3, "ls":":", "marker":"s", "ms":10, "mfc":"c"}

plt.plot([1, 4, 9, 16], **style1)
plt.plot([9, 16, 4, 1], **style2)
plt.show()
````

## 범례
여러개의 라인 플롯을 동시에 그리는 경우에는 각 선이 무슨 자료를 표시하는지를 보여주기 위해 legend명령으로 범례를 추가 가능  
범례의 위치는 자동으로 정해지지만 수동으로 설정하고 싶으면 loc키워드 인수 사용

````python
|loc|int|loc|int|
````

각 plot마다 label키워드 인수를 통해 legend label을 설정하고 legend()메서드에 loc키워즈 인자에 2을 전달함으로써 좌측 상단에 legend 위치하라고 명시

````python
x = np.linspace(-np.pi, np.pi, 256)
c, s = np.cos(x), np.sin(x)
plt.title("legend를 표시한 플롯")
plt.plot(x, c, ls = "--", label= "cosine")
plt.plot(x, s, ls = ":", label= "sine")
plt.legend(loc = 2)
plt.show()
````

## x, y축 label

# Figure

Matplotlib는 data를 Figure에 x-y 좌표의 그래프로 표시 ( 3차원은 x-y-z, 극좌표는 세타-r 등)  
각각의 Figure는 하나 이상의 Axes를 갖음  
  
간단하게 Figure, Axes객체를 만드는 방법은 pyplot.subplots()을 사용하는 것  
그 후 Axes.plot()를 활용하여 Axes에 데이터를 그림  

````python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4,], [1, 4, 2, 3])
````

figure는 전체 그림을 의미하는 객체   
figure 객체는 Axes, 특수 Artist그룹(title, figure legend, colorbar, etc), 그리고 심지어 중첩된 subfigures까지 포함 가능  
가장 쉬운 방법으로 Figure 객체를 생성하는 방법들  
Axes 없이 Figure만 얻으려면 figure() 메서드를 사용  
1개의 Axes 객체를 갖는 Figure를 만들기 위해선 subplots() 메서드를 인수 없이 사용하면 됨  
grid 형태의 shape을 갖는 Axes객체를 여럿 만들 땐 subplot() 메서드에 첫 인수로 row를 두번째 인수로 coiumn 을 입력

# Axes

Figure 객체 안에 여러 개의 플롯을 그리드 형태로 보여야하는 경우도 있음    
Figure 안에 있는 각각의 플롯은 Axes라고 불리는 객체에 속함     
Figure 안에 Axes 생성하려면 subplot()사용하려 명시적으로 Axes객체를 얻어야함   
but plot()을 바로 사용해도 자동으로 Figure, Axes 객체를 생성  

- subplot()은 그리드 형태의 Axes 객체를 생성  
- Figure가 행렬이고 Axes가 행렬의 원소

````python
ax1 = plt.subplot(2, 1, 1)
ax2 = plt.subplot(2, 1, 2)
````
